```python
def get_page_review(url):
    soup = getpage(url)
    for review_box in soup.find_all('div', class_='review-container'):
        if(type(review_box.find('span', class_='rating-other-user-rating')) != type(None)):
            rating.append(
                int(review_box.find('span', class_='rating-other-user-rating').span.text))
        else:
            rating.append(None)
        review_txt.append(review_box.find(
            'div', class_=re.compile('text')).text)
        rev_date.append(review_box.find('span', class_='review-date').text)
```
